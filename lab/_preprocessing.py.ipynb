```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
example_img = '../dataset/melanoma/ISIC_0000141.jpg'
# Loads image in from the set image path
img = tf.keras.preprocessing.image.load_img(example_img, target_size=(64, 64))
img_tensor = tf.keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
# Allows us to properly visualize our image by rescaling values in array
img_tensor /= 255.

#Plots image
plt.figure(figsize=(3,3))
plt.imshow(img_tensor[0])
plt.show()
```

```python
datagen = ImageDataGenerator(
    horizontal_flip=True,
    )

# Creates our batch of one image
pic = datagen.flow(img_tensor, batch_size=1)
plt.figure(figsize=(5, 5))

# Plots our figures
for i in range(1, 3):
    plt.subplot(1, 2, i)
    batch = pic.next()
    image_ = batch[0].astype('uint8')
    plt.imshow(image_)
plt.show()
```

```python
datagen = ImageDataGenerator(
    vertical_flip=True,
    )

# Creates our batch of one image
pic = datagen.flow(img_tensor, batch_size=1)
plt.figure(figsize=(5, 5))

# Plots our figures
for i in range(1, 3):
    plt.subplot(1, 2, i)
    batch = pic.next()
    image_ = batch[0].astype('uint8')
    plt.imshow(image_)
plt.show()
```

```python
datagen = ImageDataGenerator(
    rotation_range=30)

# Creates our batch of one image
pic = datagen.flow(img_tensor, batch_size=1)
plt.figure(figsize=(5, 5))

# Plots our figures
for i in range(1, 3):
    plt.subplot(1, 2, i)
    batch = pic.next()
    image_ = batch[0].astype('uint8')
    plt.imshow(image_)
plt.show()
```

```python
from skimage import exposure

example_img = 'dataset/vascular/ISIC_0024662.jpg'

img = tf.keras.preprocessing.image.load_img(example_img, target_size=(64, 64))
img_tensor = tf.keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)

def clahe_(img):
    img = img / 255.
    img = exposure.equalize_adapthist(img, clip_limit=0.03)
    img = img * 255.
    return img

datagen = ImageDataGenerator(
    preprocessing_function=clahe_)

pic = datagen.flow(img_tensor, batch_size=1)
plt.figure(figsize=(5, 5))

# Plots our figures
for i in range(1, 3):
    plt.subplot(1, 2, i)
    batch = pic.next()
    image_ = batch[0].astype('uint8')
    plt.imshow(image_)
plt.show()
```
